```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
import seaborn as sns
```

```python
df_sent = pd.read_csv("C:\\Users\\1221\\Desktop\\KIFYA SCHOLASHIP\\WEEK 1\\Sentiment-Analysis\\notebooks\\final sentiment data.csv")
df_sent
```

```python
selected_companies = ['AAPL', 'AMZN', 'GOOGL','GOOG', 'META','MSFT','NVDA','TSLA']
filtered_s_df = df_sent[df_sent['stock'].isin(selected_companies)]
filtered_s_df
```

```python
# Get the earliest date
start_date = filtered_s_df.groupby('stock')['date'].min()

# Get the latest date
end_date = filtered_s_df.groupby('stock')['date'].max()

print(("Start Date:", start_date, "End Date:", end_date))
```

```python
df_stock = pd.read_csv("C:\\Users\\1221\\Desktop\\KIFYA SCHOLASHIP\\WEEK 1\\Sentiment-Analysis\\notebooks\\financial stock data")

df_stock
```

```python
#renaming
#df_stock['Company'] = df_stock['Company'].replace({'APPLE': 'AAPL', 'Amazon': 'AMZN', 'Google': 'GOOG', 'Nivida': 'NVDA'})
```

```python
selected_companies = ['AAPL', 'AMZN', 'GOOG','NVDA',]
filtered_df = df_stock[df_stock['stock'].isin(selected_companies)]

filtered_df
```

```python
merged_df = pd.merge(filtered_s_df, filtered_df, on='Date', how='inner')
```

```python
merged_df.info()
```

```python

sentiment_category = merged_df['sentiment_category'].value_counts()

# Plotting the distribution of entries by company
plt.figure(figsize=(8, 6))
sentiment_category.plot(kind='bar', color='skyblue')
plt.title('Distribution of Entries by sentiment category')
plt.xlabel('sentiment category')
plt.ylabel('Number of Entries')
plt.xticks(rotation=45)
plt.show()
```

```python
merged_df['Daily_Return'] = merged_df['Close'].pct_change() * 100  # Returns in percentage
merged_df
```

```python
# plot distribution with different colors
plt.figure(figsize=(10, 6))

sns.histplot(merged_df['Daily_Return'], kde=True, color='blue')

plt.title('Distribution of Daily Return')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()
```

```python
# Assume merged_df is your DataFrame
selected_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'sentiment score', 'Daily_Return']
df_selected = merged_df[selected_columns]

# Calculate the correlation matrix
corr_matrix = df_selected.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of General Sentiment Category')
plt.show()
```

```python
selected_Neg = merged_df[merged_df['sentiment_category'].isin(['Negative'])]
```

```python
# Assume df is your DataFrame
selected_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'sentiment score', 'Daily_Return']
df_selected = selected_Neg[selected_columns]

# Calculate the correlation matrix
corr_matrix = df_selected.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', linewidths=0.5)
plt.title('Correlation Heatmap of Negative Sentiments')
plt.show()
```

```python
selected_pos = merged_df[merged_df['sentiment_category'].isin(['Positive'])]
```

```python
# Assume df is your DataFrame
selected_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'sentiment score', 'Daily_Return']
df_selectedP = selected_pos[selected_columns]

# Calculate the correlation matrix
corr_matrix = df_selectedP.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='plasma', linewidths=0.5)
plt.title('Correlation Heatmap of Positive Sentiments')
plt.show()
```

```python
selected_Neutral = merged_df[merged_df['sentiment_category'].isin(['Neutral'])]
```

```python
# Assume df is your DataFrame
selected_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'sentiment score', 'Daily_Return']
df_selectedNew = selected_Neutral[selected_columns]

# Calculate the correlation matrix
corr_matrix = df_selectedNew.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='magma', linewidths=0.5)
plt.title('Correlation Heatmap of Neutral Sentiments')
plt.show()
```
